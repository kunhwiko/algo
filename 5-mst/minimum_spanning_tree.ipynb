# Minimum Spanning Tree

## Part 1: Minimum Spanning Tree

**Explanation**

The goal of connecting cities at a minimum total cost. 

<br />

**Assumption**

The graph is connected and undirected.

<br />

**Terminology**

Assume "A" is a set of edges that is part of the MST. Now how would you grow from A to the MST one edge at a time? 

<br />

cut: a partition of a vertex set S and V-S 

* edge is said to "cross-cut" S and V-S if one point is in S and the other is in V-S 

* cut (S, V-S) "respects" the set of edges A if each edge in A is not crossing the cut 

light edge: edge of minimum weight that crosses cut 

Screen Shot 2021-01-05 at 4.59.57 PM.png

<br />

**Theorem**

The MST theorem states that if A (set of edges within S) is a part of the MST and (S,V-S) is a cut that respects A, then it must be safe to add a light edge to A to grow into the MST.

## Part 2: Prim's Algorithm 

**Explanation**

Assuming there is a set of edges A that is always a tree, look for the light edge that grows from a vertex in this tree to a vertex that is not part of this tree.

This algorithm is similar to Dijkstra's, as can be seen in the pseudocode.

<br />

**Pseudocode**

initialize shortest and pred 

shortest[src] = 0, the rest is Inf

pred is all null

put all vertices in Q 

<br />

while Q is not empty:

&nbsp;&nbsp;&nbsp;&nbsp;u = vertex with minimum shortest value and in Q

&nbsp;&nbsp;&nbsp;&nbsp;delete u in Q 

&nbsp;&nbsp;&nbsp;&nbsp;for v in adj(u):

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if v in Q and shortest[v] > wt(u,v)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;shortest[v] = wt(u,v)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;pred[v] = u 

<br />

**Runtime**

Array Based Implementation: $O(V^2)$

Heap Based Implementation: $O((V+E)logV)$

## Part 3: Kruskal's Algorithm

### Part 3.1 Kruskal's Overview 

**Explanation**

Process of selecting edges in increasing weight order to create a MST. 

<br />

**Operations**

* makeset(x): makes a singleton disjoint set {x} if x is not part of any other set 

* union(x,y): figure out which sets x and y are part of, then unfiy the two sets 

* findset(x): return a pointer to the representative of the set that contains x 

<br />

**Pseudocode**

Initialize A as an empty set (A is the set of edges that form the MST)

for each v in V:

&nbsp;&nbsp;&nbsp;&nbsp;makeset(v)

sort edges in increasing weight order

for (u,v) in sorted E:

&nbsp;&nbsp;&nbsp;&nbsp;if findset(u) != findset(v)

&nbsp;&nbsp;&nbsp;&nbsp;add (u,v) to A 

&nbsp;&nbsp;&nbsp;&nbsp;union(u,v)

return A 

<br />

**Runtime**

V*makeset + ElogE + O(E) * (findset + union)

$\therefore$ $O(ElogE)$ (details on makeset, findset, union explained below)
